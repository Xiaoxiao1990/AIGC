一些概念

**ReLU**: 
全称为：Rectified Linear Unit，是一种人工神经网络中常用的激活函数，通常意义下，其指代数学中的斜坡函数，即
$$ f(x) = max(0, x) $$

```python
import numpy as np
import matplotlib.pyplot as plt

# plot ReLU function
x = np.linspace(-3, 3, 3)
y = np.maximum(0, x)
plt.plot(x, y, 'r')
```
