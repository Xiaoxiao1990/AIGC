```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
def f(x, y):
   return np.sin(np.sqrt(x ** 2 + y ** 2))
#构建x、y数据
x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)
#将数据网格化处理
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
fig = plt.figure()
ax = plt.axes(projection='3d')
#50表示在z轴方向等高线的高度层级，binary颜色从白色变成黑色
ax.contour3D(X, Y, Z, 50)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('3D contour')
plt.show()
```

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
#要绘制函数图像
def f(x, y):
   return np.sin(np.sqrt(x ** 2 + y ** 2))
#准备x,y数据
x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)
#生成x、y网格化数据
X, Y = np.meshgrid(x, y)
#准备z值
Z = f(X, Y)
#绘制图像
fig = plt.figure()
ax = plt.axes(projection='3d')
#调用绘制线框图的函数plot_wireframe()
ax.plot_wireframe(X, Y, Z)
ax.set_title('wireframe')
plt.show()
```

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
#求向量积(outer()方法又称外积)
x = np.outer(np.linspace(-2, 2, 30), np.ones(30))
#矩阵转置
y = x.copy().T 
#数据z
z = np.cos(x ** 2 + y ** 2)
#绘制曲面图
fig = plt.figure()
ax = plt.axes(projection='3d')
#调用plot_surface()函数
ax.plot_surface(x, y, z,cmap='viridis', edgecolor='none')
ax.set_title('Surface plot')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# (x - a)^2 + (y - b)^2 = r^2
def draw_circle(x0, y0, r):    
    _ , axes = plt.subplots()
    scale = np.pi * 2
    dot_num = 200
    u = np.linspace(0, scale, dot_num)
    x = x0 + r * np.cos(u)
    y = y0 + r * np.sin(u)
    # 设置坐标轴x/y 显示比例
    axes.set_aspect(1)
   
    axes.plot(x,y)
   
    axes.scatter(x0, y0)

    dot_idx = 25

    xt,yt = x0 + r * np.cos(u[dot_idx]), y0 + r * np.sin(u[dot_idx])

    axes.scatter(xt, yt)

    xs = np.linspace(x0, xt)
    ys = np.linspace(y0, yt)
    axes.plot(xs,ys, color ='blue', linewidth=1, linestyle="--")
    
    axes.text(x0, y0, (x0, y0))
    axes.text(xt, yt, (xt, yt))

# draw_circle(0,1,1)

def fmt(x):
    s = f"{x:.1f}"
    if s.endswith("0"):
        s = f"{x:.0f}"
    return rf"{s} \%" if plt.rcParams["text.usetex"] else f"{s} %"

def draw_ball(x0, y0, r):    
    #绘制一张图
    fig=plt.figure()
    #绘制3D子图
    ax=fig.add_subplot(projection="3d")
    # u = np.arange(-3,3,0.1)
    # v = np.arange(-3,3,0.1)

    u = np.linspace(0, 2*np.pi, 100)
    v = np.linspace(0, np.pi, 100)

    x,y= np.meshgrid(np.cos(u), np.sin(v))
    z = np.array(x**2 + y**2)
    
    # ax.plot_surface(x0,x1,z,cmap='inferno')

    #rstride表示行步长，cmap给曲面加上色彩
    surf = ax.plot_surface(x,y,z,rstride=1,cstride=1,cmap="inferno")
    #colorbar可以给图像添加一个颜色条
    fig.colorbar(surf)
    CS = ax.contour(x,y,z)
    ax.clabel(CS, CS.levels, inline=True, fmt=fmt, fontsize=10)
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.set_zlabel("z")
    ax.set_title("3D")
    plt.show()


draw_ball(0,0,0)
```

```python

import matplotlib.pyplot as plt
import numpy as np
 
# def f(x,y):
#     return (1-x/2+x**5+y**3)*np.exp(-x**2-y**2)
 
def f(x,y):
    return (1-x/2+x**4+y**3)*np.exp(-x**2-y**2)

n = 256
x = np.linspace(-3,3,n)
y = np.linspace(-3,3,n)
 
# 生成网格
X,Y = np.meshgrid(x,y)
 
# contourf 会在等高线之间进行填充
plt.contourf(X,Y,f(X,Y),8,alpha=0.75,cmap='hot')
 
# contour 仅仅是画出等高线
C = plt.contour(X,Y,f(X,Y),8,colors='black',linewidth=.4)
 
# 添加数据标签
plt.clabel(C,inline=True,fontsize=10)
 
plt.xticks(())
plt.yticks(())
 
```
