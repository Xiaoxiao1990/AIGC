准备好了就可以试一下了。试着画一个$y = x$的图像。首先通过numpy产生一个数组，然后让y等于x。再用matplot画出图像。

```python
import numpy as np;
import matplotlib.pyplot as plt;

x = np.arange(10);
y = x;
plt.plot(x, y)
```

以上就是它的图像了。但是为了画出更复杂的图像，如一元二次函数，从数学知识可知 $f(x)=x^2$ ，那就要让数组y中的元素分别对 $x^2$ 对应起来。这个思路应该没问题，那么，现在就要了解怎么操作numpy数组了。通过网上找资料知道，numpy有一个nditer操作，它是一个迭代器，可以遍历整个数组。那么可以试试。

```python
i = 0;
for xi in np.nditer(x):
    y[i] = xi * xi;
    i = i + 1;

plt.plot(x, y)
```

看上去是可行的。这里还有一点要知道的是，数组的索引是用中括号“[]”及下标表示的。下标起始点为0，即数组的第一个元素是0。但是还有一个更简单的方法，因为numpy自带了许多运算操作，其中就有power()运算。因此，可以更简单地实现如下：

```python
y = np.power(x,2) # 即，将x中的每个元素都做二次方运算，然后得到一个新的数组y
plt.plot(x, y)
```
