# 绘制图形

球的标准方程
$$x^2+y^2+z^2=R^2$$

转为参数方程$$x=rsin\varphi cos\theta \\ y=rsin\varphi sin\theta \\ z=rcos\varphi \\(其中0\le \theta \le 2\pi,0\le \varphi \le \pi)$$



```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.animation as animation

def draw_ball(x0,y0,z0,r):
    fig = plt.figure()
    ax = fig.add_subplot(projection='3d')

    u = np.linspace(0, 2 * np.pi, 100)
    v = np.linspace(0, np.pi, 100)

    x = r * np.outer(np.cos(u), np.sin(v))
    y = r * np.outer(np.sin(u), np.sin(v))
    z = r * np.outer(np.ones(np.size(u)), np.cos(v))

    ax.plot_surface(x,y,z)

def draw_line(x1,y1,x2,y2):
    plt.axhline(y=0, c='black')
    plt.axvline(x=0, c='black')
    plt.plot([x1,x2],[y1,y2])

# draw_line(0,0,2,2)

def draw_dot(x,y):
    plt.axhline(y=0, c='black')
    plt.axvline(x=0, c='black')
    plt.scatter(x,y)

# x = [1,2,3,4]
# y = [1,2,3,4]
# X,Y = np.meshgrid(x,y)

# print(X)
# print(Y)
# c =X+Y
# plt.scatter(X,Y,c=c)
# plt.plot(X,Y)
# plt.plot(Y,X)


fig = plt.figure()
ax = fig.add_subplot(projection='3d')

x = np.linspace(0, 6*np.pi, 100)
y = np.sin(x)
# z = np.zeros(np.size(x))
# z[:] = 1
z = np.cos(x)


ax.plot(x, y,z)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```
