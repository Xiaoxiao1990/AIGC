```python
team_num = 9
members_per_team = 3

def create_team(team_num):
    return [i for i in range(1, team_num+1)]

array = create_team(team_num)

print("array: ", array)

ranked = [0 for i in range(team_num*members_per_team)]

def rank_pos(ranked, pos, team_idx):
    if pos >= len(ranked):
        return -1
    
    m1 = pos
    m2 = m1 + team_idx + 1
    if m2 >= len(ranked):
        return -1

    m3 = m2 + team_idx + 1
    if m3 >= len(ranked):
        return -1
    
    if (ranked[m1] != 0) or (ranked[m2] != 0) or (ranked[m3] != 0):
        return -2

    ranked[m1] = team_idx
    ranked[m2] = team_idx
    ranked[m3] = team_idx

    return 0

def get_ranked_blank_pos(ranked):
    for i in range(len(ranked)):
        if ranked[i] == 0:
            return i
    return -1

# delete team_idx from array
def delete_team(array, team_idx):
    for i in range(len(array)):
        if array[i] == team_idx:
            array[i] = 0
            return 0
    return -1

# get rest team in array
def get_rest_team(array):
    rest_team = []
    for i in range(len(array)):
        if array[i] != 0:
            rest_team.append(array[i])
    return rest_team

def rank(ranked, array):
    rest_team = get_rest_team(array)
    if len(rest_team) == 0:
        # print("rank: ", ranked)
        return 0
    
    pos = get_ranked_blank_pos(ranked)
    if pos == -1:
        return 0
    
    for i in range(len(rest_team)):
        team_idx = rest_team[i]
        if rank_pos(ranked, pos, team_idx) == 0:
            delete_team(array, team_idx)
            if rank(ranked, array) == 0:
                print("rank: ", ranked)
                # return 0
            else:
                ranked[pos] = 0
                ranked[pos+team_idx+1] = 0
                ranked[pos+team_idx+1+team_idx+1] = 0
                array.append(team_idx)

rank(ranked, array)

```
